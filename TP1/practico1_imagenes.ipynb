### 0. Lectura de imagen y datos generales

```python
im = imageio.imread('koala.jpg')
print("Alto: ",im.shape[0])
print("Ancho: ",im.shape[1])
print("Profundidad: ",im.shape[2])
if (im.shape[2] == 3):
  print("La imagen leída es RGB")
```

```python
print(im[0])
print(im.max())
print(im.min())
print(im.mean())
print(type(im))
```

La imagen se representa con 8 bits, siendo su valor mínimo 0 y su valor máximo 255.

### 1. Mostrar la imagen, realizar cambios en un canal y mostrarla de vuelta
En esta sección se reemplazarán los valores de los píxeles del canal R por cero, haciendo que la imagen cambie significativamente ante la ausencia de intensidad en dicho canal.

```python
plt.figure(figsize = (8,8))
plt.imshow(im)
plt.show()
```

```python
for i in range(len(im)):
  for j in range(len(im[0])):
    im[i][j][0] = 0 
```

```python
plt.figure(figsize = (8,8))
plt.imshow(im)
plt.show()
```

### Guardamos la nueva imagen en un archivo

```python
imageio.imwrite("koala_noR.jpg",im)
```

```python
im_noR = imageio.imread('koala_noR.jpg')
plt.figure(figsize = (8,8))
plt.imshow(im_noR)
plt.show()
```
